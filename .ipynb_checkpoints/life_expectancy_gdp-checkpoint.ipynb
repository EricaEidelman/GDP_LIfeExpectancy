```python
# Import dependencies
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in data
file = "all_data.csv"
census_df = pd.read_csv(file)

census_df.head()
```

```python
# Find datatypes
census_df.dtypes
```

```python
# Summarize countries and years in data
countries = census_df["Country"].tolist()
country = []
for item in countries:
    if item not in country:
        country.append(item)
        
years = census_df["Year"].tolist()
year = []
for item in years:
    if item not in year:
        year.append(item)
        
print(f"The dataset includes GDP and life expectancies on the following countries - {country} - for the years {year[0]} through {year[len(year)-1]}.")
```

```python
# Find min and max GDP values
min = census_df.GDP.min()
print(min)

max = census_df.GDP.max()
print(max)
```

```python
# Add column showing GDP converted to GDP in trillions of USD
census_df["GDP_USD_Trillions"] = round(census_df["GDP"]/1000000000000,2)
census_df.head()
```

```python
# Rename Life Expectancy column to a more manageable name
census_df.rename(columns = {"Life expectancy at birth (years)":"Life_Expectancy"}, inplace = True)
census_df.head()
```

```python
# Obtain overall summary statistics
max_gdp = census_df["GDP_USD_Trillions"].max()
min_gdp = census_df["GDP_USD_Trillions"].min()
mean_gdp = census_df["GDP_USD_Trillions"].mean()
max_life = census_df["Life_Expectancy"].max()
min_life = census_df["Life_Expectancy"].min()
mean_life = census_df["Life_Expectancy"].mean()
print(f"The maximum GDP in USD trillions is {max_gdp}. The minimum is {min_gdp}. The average is {round(mean_gdp,2)}.")
print(f"The maximum life expectancy in years is {max_life}. The minimum is {min_life}. The average is {round(mean_life,2)}.")
```

```python
# Obtain summary statistics by country.
for item in country:
    max_gdp = census_df["GDP_USD_Trillions"][census_df["Country"] == item].max()
    min_gdp = census_df["GDP_USD_Trillions"][census_df["Country"] == item].min()
    mean_gdp = census_df["GDP_USD_Trillions"][census_df["Country"] == item].mean()
    max_life = census_df["Life_Expectancy"][census_df["Country"] == item].max()
    min_life = census_df["Life_Expectancy"][census_df["Country"] == item].min()
    mean_life = census_df["Life_Expectancy"][census_df["Country"] == item].mean()
    print(f"The maximum GDP in USD trillions for {item} is {max_gdp}. The minimum is {min_gdp}. The average is {round(mean_gdp,2)}.")
    print(f"The maximum life expectancy in years for {item} is {max_life}. The minimum is {min_life}. The average is {round(mean_life,2)}.")
```

```python
# Split df into dfs based on country
chile = census_df[census_df["Country"] == "Chile"]
china = census_df[census_df["Country"] == "China"]
germany = census_df[census_df["Country"] == "Germany"]
mexico = census_df[census_df["Country"] == "Mexico"]
usa = census_df[census_df["Country"] == "United States of America"]
zimbabwe = census_df[census_df["Country"] == "Zimbabwe"]
```

```python
# Plot comparison of GDP across the countries
plt.plot(chile["Year"], chile["GDP_USD_Trillions"])
plt.plot(china["Year"], china["GDP_USD_Trillions"])
plt.plot(germany["Year"], germany["GDP_USD_Trillions"])
plt.plot(mexico["Year"], mexico["GDP_USD_Trillions"])
plt.plot(usa["Year"], usa["GDP_USD_Trillions"])
plt.plot(zimbabwe["Year"], zimbabwe["GDP_USD_Trillions"])
plt.title("Country GDP (USD Trillions)")
plt.legend(["Chile", "China", "Germany", "Mexico", "USA", "Zimbabwe"], bbox_to_anchor = (1, 1), loc = "upper left")
plt.show()
```

```python
# Plot comparison of life expectancy across the countries
plt.plot(chile["Year"], chile["Life_Expectancy"])
plt.plot(china["Year"], china["Life_Expectancy"])
plt.plot(germany["Year"], germany["Life_Expectancy"])
plt.plot(mexico["Year"], mexico["Life_Expectancy"])
plt.plot(usa["Year"], usa["Life_Expectancy"])
plt.plot(zimbabwe["Year"], zimbabwe["Life_Expectancy"])
plt.title("Country Life Expectancy")
plt.legend(["Chile", "China", "Germany", "Mexico", "USA", "Zimbabwe"], bbox_to_anchor = (1, 1), loc = "upper left")
plt.show()
```

```python
plt.scatter(chile["Life_Expectancy"], chile["GDP_USD_Trillions"])
plt.xlabel("Life Expectancy")
plt.ylabel("GDP (USD Trillions)")
plt.title("Life Expectancy vs. GDP (Chile)")
plt.show()
```
